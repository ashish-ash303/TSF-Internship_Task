## Importing Libraries


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Importing Dataset

```python
df=pd.read_csv("SampleSuperstore.csv")
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Sub-Category"].unique()
```

```python
store_df=df.drop(["Country","State","Postal Code"],axis=1)
store_df
```

## Finding Insights for South Region

```python
South_df=store_df.query('Region == "South"')
South_df
```

```python
South_df.shape
```

> * Our Dataset has 1620 rows of South Region 

## Q.We start  Looking with Which Category and Sub-Category has  highest Sale in South region

```python
Sales_df=South_df.sort_values("Sales",ascending=False)
Sales_df
```

```python
plt.figure(figsize=(14,10))
plt.grid(color='g', linestyle='-', linewidth=0.6,alpha=0.6)
plt.xlabel("Category",fontsize=15)
plt.tick_params(axis='x', which='major', labelsize=15)
plt.ylabel("Sales",fontsize=15)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.title("Category vs Sales",fontsize=15,pad=19)
plt.bar(Sales_df.Category,Sales_df.Sales);
```

### *In South region, Category Technology and  sub Category Machines Has Highest Sales* 

```python
lowest_Sales_df=South_df.sort_values("Sales",ascending=True)
lowest_Sales_df
```

```python
lowest_Sales_df.Category.unique()
```

```python
plt.figure(figsize=(14,10))
plt.grid(color='g', linestyle='-', linewidth=0.6,alpha=0.6)
plt.xlabel("Category",fontsize=15)
plt.tick_params(axis='x', which='major', labelsize=15)
plt.ylabel("Sales",fontsize=15)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.title("Category vs Sales",fontsize=15,pad=19)
plt.bar(lowest_Sales_df.Category,lowest_Sales_df.Sales);
```

### *In South region, Category Furniture Has Lowest Sales*

## Q2 Which Segment  has highest Profit in South region

```python
profit_df=South_df.sort_values("Profit",ascending=False)
profit_df
```

```python
plt.figure(figsize=(14,10))
plt.grid(color='g', linestyle='-', linewidth=0.6,alpha=0.6)
plt.xlabel("Category",fontsize=15)
plt.tick_params(axis='x', which='major', labelsize=15,rotation=90)
plt.ylabel("Profit",fontsize=15)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.title("Segment vs Profit",fontsize=15,pad=19)
plt.bar(profit_df.Segment,profit_df.Profit,color=(0.2, 0.4, 0.6, 0.6));
# sns.barplot(x = 'Category',  y ='Profit', data = profit_df);
```

## *Consumer Segment Has highest Profit in South region*

## Q. Which Segment has highest loss in South Region 

```python
lowest_profit_df=South_df.sort_values("Profit",ascending=True)
lowest_profit_df
```

```python
plt.figure(figsize=(14,10))
plt.grid(color='g', linestyle='-', linewidth=0.6,alpha=0.6)
plt.xlabel("Category",fontsize=15)
plt.tick_params(axis='x', which='major', labelsize=15,rotation=90)
plt.ylabel("Profit",fontsize=15)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.title("Segment vs Profit",fontsize=15,pad=19)
plt.bar(lowest_profit_df.Segment,lowest_profit_df.Profit,color=(0.2, 0.4, 0.6, 0.6));
```

## *Corporate Segment Has Highest Loss*

## Q. Which Category Has highest Quantity Sale

```python
Quantity_sales_df=South_df.sort_values("Quantity",ascending=False)
Quantity_sales_df
```

```python
plt.figure(figsize=(14,10))
plt.grid(color='g', linestyle='-', linewidth=0.6,alpha=0.6)
plt.xlabel("Category",fontsize=15)
plt.tick_params(axis='x', which='major', labelsize=15,rotation=90)
plt.ylabel("Quantity",fontsize=15)
plt.tick_params(axis='y', which='major', labelsize=15)
plt.title("Category vs Quantity",fontsize=15,pad=19)
plt.bar(Quantity_sales_df.Category,Quantity_sales_df.Quantity,color=(0.2, 0.4, 0.6, 0.6));
```
